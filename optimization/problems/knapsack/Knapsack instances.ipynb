## Convert raw data file to qudratic program 

```python
from qiskit_optimization.applications import Knapsack
```

```python

def read_data(filename):
    data_file = open(filename, 'r')
    count = 0
    p=[]
    w=[]
    while True:
        count += 1
        line=data_file.readline()
        if not line:
            break
        dp = [int(i) for i in line.rstrip('\n').split(' ')]
        if (count == 1):
            n=dp[0]
            b=dp[1]
        else:
            p.append(dp[0])
            w.append(dp[1])
    data_file.close()
    prob=Knapsack(p,w,b)
    qp=prob.to_quadratic_program()
    return qp 
```

```python
qp = read_data('./kp5/kp_5_80')
print(qp.export_as_lp_string())
```
